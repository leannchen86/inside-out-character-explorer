```python
import numpy as np

def plot_network_snapshot(interaction_data_all_scenes, scene_number):
    # Extracting data for the specific scene
    interaction_data = interaction_data_all_scenes[scene_number]
    
    if not interaction_data:  # Skip if the scene has no interactions
        print(f"No interactions in scene {scene_number + 1}. Skipping...")
        return
    
    G = nx.Graph()

    # Min and max values for normalization
    min_sentiment = sentiment_values(interaction_data_all_scenes, is_all_scenes=True)[0]
    max_sentiment = sentiment_values(interaction_data_all_scenes, is_all_scenes=True)[1]

    for interaction, data in interaction_data.items():
        char1, char2 = list(interaction)
        count = data['count']
        sentiment = data['sentiment']
        
        # Normalizing sentiment between 0 and 1 for color mapping
        normalized_sentiment = (sentiment - min_sentiment) / (max_sentiment - min_sentiment)
        
        G.add_edge(char1, char2, weight=count, sentiment=normalized_sentiment)

    # Calculating eigenvector centrality if the graph has more than one node
    if len(G) > 1:
        try:
            centrality = nx.eigenvector_centrality_numpy(G, weight='weight')
        except:
            centrality = {node: 1 for node in G.nodes()}  # Default centrality
    else:
        centrality = {node: 1 for node in G.nodes()}  # Default centrality
    
    # Creating the plot
    # Creating the plot
    pos = nx.spring_layout(G, seed=42)

    # If centrality could not be calculated, default to a single color for all nodes
    if all(value == 1 for value in centrality.values()):
        node_color = 'blue'
    else:
        node_color = [centrality[node] for node in G.nodes()]

    nx.draw(G, pos, 
            with_labels=True, 
            node_color=node_color,
            edge_color=[G[u][v]['sentiment'] for u, v in G.edges()],
            edge_cmap=plt.cm.RdYlGn,
            cmap=plt.cm.viridis)

    plt.title(f'Network Snapshot for Scene {scene_number + 1}')
    plt.show()
```

```python
from collections import defaultdict

def extract_weighted_interactions(scenes, characters):
    interactions = defaultdict(float)
    
    for scene in scenes:
        for i in range(len(characters)):
            char1 = characters[i]
            for j in range(i+1, len(characters)):
                char2 = characters[j]
                if char1 in scene and char2 in scene:
                    # Count the frequency of each character in the scene
                    count_char1 = scene.count(char1)
                    count_char2 = scene.count(char2)
                    
                    # Calculate the interaction weight as the sum of their frequencies
                    interaction_weight = count_char1 + count_char2
                    
                    interactions[(char1, char2)] += interaction_weight / 2
                    interactions[(char2, char1)] += interaction_weight / 2
                    
    return interactions

# Extract Joy and Sadness interactions from the scenes
joy_sadness_interactions = extract_weighted_interactions(splits, ['JOY', 'SADNESS'])
joy_sadness_interactions
```
